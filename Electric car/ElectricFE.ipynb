```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the plot style
sns.set(style="whitegrid")

# Group the data by City and Electric Vehicle Type and calculate the count
grouped_data = df.groupby(['City', 'Electric Vehicle Type']).size().reset_index(name='Count')

# Create a line plot
plt.figure(figsize=(12, 6))
sns.lineplot( x='City',y='Count', hue='Electric Vehicle Type', data=grouped_data, marker='o')

# Set the title and labels
plt.title('Distribution of Electric Vehicle Types in Cities')
plt.xlabel('Cities')
plt.ylabel('Count')

# Rotate the x-axis labels for better readability
plt.xticks()

# Display the plot
plt.show()
```
