```python
df = pd.read_json("pokedex.json")
data=pd.DataFrame(df.pokemon.values.tolist())
data
```

```python
data["img"]=data["img"].astype("string")
data["type"]=data["type"].astype("string")
data["candy"]=data["candy"].astype("string")
data["num"]=data["num"].astype("int64")
data["name"]=data["name"].astype("string")
```

```python
data.info()
```

```python
data["height"].unique()
```

```python
data["height"]=data["height"].str.replace("m","")
```

```python
data["height"]=data["height"].astype("float64")
```

```python
data.info()
```

```python
data["weight"].unique()
```

```python
data["weight"]=data["weight"].str.replace("kg"," ")
```

```python
data["weight"]=data["weight"].astype("float64")
```

```python
data["candy_count"].unique()
```

```python
data["candy_count"]=data["candy_count"].fillna(value=data["candy_count"].mean())
```

```python
data["candy_count"]=data["candy_count"].astype("int64")
```

```python
data["egg"]=data["egg"].replace("Not in Eggs","0")
```

```python
##According to Pokémon Go's website, 7 kilometer eggs are now only hatching fossils. Specifically, the Pokémon players can find from these eggs are Omanyte, Kabuto, Aerodactyl, Lileep, Anorith, Cranidos, Shieldon, Tirtouga, and Archen. These Pocket Monsters will no longer hatch from any other eggs, which means if players want to obtain one, they need a friend to send them a gift and hope it contains a 7 km egg.
data["egg"]=data["egg"].str.replace("Omanyte Candy","7")
```

```python
data["egg"]=data["egg"].str.replace("km"," ")
```

```python
data["egg"]=data["egg"].astype("float64")
```

```python
data["spawn_chance"]= data["spawn_chance"].round(2)
```

**Question 6 Draw plots to demonstrate the analysis for the following questions for better visualizations.**

Get all Pokemons that have less than 4 weaknesses

```python
import pandas as pd
import matplotlib.pyplot as plt


ploted_data = data[data['weaknesses'].apply(len) < 4]

ploted_data['weaknesses'].apply(len).value_counts().sort_index().plot(kind='bar')
plt.xlabel('Number of Weaknesses')
plt.ylabel('Count')
plt.title('Pokemon Weakness Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

evolution_counts = 0
no_evolution_counts = 0

for row in data['next_evolution']:
    if isinstance(row, list):
        if len(row) <=2:
            evolution_counts += 1
        else:
            no_evolution_counts += 1

labels = ['Pokmon with not more than 2 Evolutions ', 'Pokemon with more than 2 Evolutions']
counts = [evolution_counts, no_evolution_counts]

colors = ['lightblue', 'lightgreen'] 
plt.pie(counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=180)
plt.title('Pokemon Evolution Distribution')
plt.axis('equal')
plt.show()
```
