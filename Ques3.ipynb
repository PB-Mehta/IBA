a. Measure the dispersion in both and interpret the results.

```python
import pandas as pd
import numpy as np
df= pd.read_csv('ques3data.csv')
```

```python
bpb = df[' Blood Pressure Before (mmHg)']
bpa = df[' Blood Pressure After (mmHg)']
```

```python
sd_bpb = np.std(bpb)
sd_bpa = np.std(bpa)
print('Standard Deviation before',sd_bpb)
print('Standard deviation afrer',sd_bpa)
print("The standard deviation measures the dispersion of the data around the mean.")
print("A higher standard deviation indicates greater variability in the data.")
```

b. Calculate mean and 5% confidence interval and plot it in a graph

```python
import matplotlib.pyplot as plt
from scipy.stats import t

mean_b =np.mean(bpb)
std_b= np.std(bpb)
n_before=len(bpb)
t_val_bef=t.ppf(0.975,n_before-1)
ci_low_before = mean_b - t_val_bef * (std_b / np.sqrt(n_before))
ci_high_before = mean_b + t_val_bef * (std_b / np.sqrt(n_before))

mean_a =np.mean(bpa)
std_a= np.std(bpa)
n_after=len(bpa)
t_val_aft=t.ppf(0.975,n_after-1)
ci_low_after = mean_a - t_val_aft * (std_a / np.sqrt(n_after))
ci_high_after = mean_a + t_val_aft * (std_a / np.sqrt(n_after))

labels = ['Before', 'After']
means = [mean_b, mean_a]
ci_lows = [ci_low_before, ci_low_after]
ci_highs = [ci_high_before, ci_high_after]

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x - width / 2, means, width, label='Mean')
rects2 = ax.errorbar(x - width / 2, means, yerr=[means[0] - ci_lows, ci_highs - means[1]], fmt='none', capsize=5,label='95% CI')

ax.set_ylabel('Blood Pressure (mmHg)')
ax.set_title('Mean and 95% Confidence Interval of Blood Pressure')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```

Calculate the Mean absolute deviation and Standard deviation and interpret
the results.

```python
mad_bef=np.mean(np.abs(bpb-np.mean(bpb)))
mad_aft=np.mean(np.abs(bpa-np.mean(bpa)))
sd_bef=np.std(bpb)
sd_aft=np.std(bpa)

print("Mean Absolute Deviation (MAD) - Before: {:.2f}".format(mad_bef))
print("Mean Absolute Deviation (MAD) - After: {:.2f}".format(mad_aft))
print("Standard Deviation (SD) - Before: {:.2f}".format(sd_bef))
print("Standard Deviation (SD) - After: {:.2f}".format(sd_aft))
```

```python
from scipy.stats import pearsonr
corr_coeff ,p_value=pearsonr(bpb,bpa)
print("Correlation Coefficient: {:.2f}".format(corr_coeff))
print("p-value: {:.4f}".format(p_value))

alpha = 0.01

if p_value < alpha:
    print("The correlation is significant at the 1% level.")
else:
    print("The correlation is not significant at the 1% level.")
```
