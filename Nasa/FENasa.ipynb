Get all the Earth meteorites that fell before the year 2000

```python
import pandas as pd
import matplotlib.pyplot as plt

# filter dates before 2000-01-01
dates = df['year'][df['year'] < pd.to_datetime('2000-01-01')]

plt.figure(figsize=(10, 6))
plt.hist(dates, bins=30, color='skyblue',edgecolor='black', alpha=0.7)
plt.title('Dates Before 2000-01-01')
plt.xlabel('Date')
plt.ylabel('Frequency')
plt.xticks(rotation=45,ha='right',fontsize = 8 )
plt.grid(axis='y', linestyle='--')
plt.show()
```

Get all the earth meteorites co-ordinates who fell before the year 1970

```python
filtered_meteorites = df[(df['year'] < pd.to_datetime('1970-01-01'))]
```

```python
coordinates = filtered_meteorites['geolocation']

plt.figure(figsize=(10, 6))
plt.scatter(coordinates.str[1], coordinates.str[0], color='blue', alpha=0.5,marker ='o',edgecolor='black')
plt.title('Earth Meteorite Landings Before 1970', fontsize=16)
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True,linestyle='--',linewidth=0.5)
plt.tight_layout()
plt.show()

```
